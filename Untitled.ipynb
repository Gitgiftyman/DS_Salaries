# DS PROJ SALARY ANALYTICS

```python
#Libraries
import pandas as pd
import numpy as np
```

```python
#read file as a data frame
df=pd.read_csv('glassdoor_jobs.csv')
df
```

```python
#get rid of null values (-1) values in the dataframe
df2=df[df['Salary Estimate'] != '-1']
df2
```

```python
salary=df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```python
remove_Ks= salary.apply(lambda x: x.replace('K','').replace('$',''))
remove_Ks
```

```python
remove_ph= remove_Ks.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
remove_ph
```

```python
df['hourly']= df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df.head(70)
```

```python
df['min_salary']= remove_ph.apply(lambda x: (x.split('-')[0]))
df['min_salary']=pd.to_numeric(df['min_salary'])
df1=df['min_salary']
df1=df1.dropna(axis=0)
df1=df1.astype(int)
df1
```

```python
df['max_salary']= remove_ph.apply(lambda x: int(x.split('-')[1]))
df['max_salary']
```

```python
df['avg_salary']=(df.max_salary +df1)/2
df['avg_salary']
```
